# Лабораторная работа 2. Понижение размерности пространства признаков

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

from sklearn import preprocessing
from sklearn.decomposition import PCA
from sklearn.decomposition import KernelPCA
from sklearn.decomposition import SparsePCA
from sklearn.decomposition import FactorAnalysis
```

## Загрузка данных

```python
df = pd.read_csv('glass.csv')
var_names = list(df.columns) # получение имен признаков
labels = df.to_numpy('int')[:,-1] # метки классов
data = df.to_numpy('float')[:,:-1] # описательные признаки
data = preprocessing.minmax_scale(data) # нормикровка данных
```

```python
def show_scatters(data):
    fig, axs = plt.subplots(2,4, figsize=(14, 7))
    for i in range(data.shape[1] - 1):
        axs[i // 4, i % 4].scatter(data[:,i], data[:,(i+1)], c=labels, cmap='hsv')
        axs[i // 4, i % 4].set_xlabel(var_names[i])
        axs[i // 4, i % 4].set_ylabel(var_names[i+1])
    plt.show()
```

```python
show_scatters(data)
```

```python
# Опрделяем соответсвие цвета на диаграмме и класса в датасете
plt.scatter(labels, [0]*len(labels), c=labels, cmap='hsv')
plt.xlabel('Класс в датасете')
plt.show()
```

## Метод главных компонент

```python
pca = PCA(n_components = 4)
pca_data = pca.fit(data).transform(data) # понижаем размерность
```

```python
def print_pca(pca):
    print('Var: ', pca.explained_variance_ratio_) #значение объясненной дисперсии
    print('Eigenvals: ', pca.singular_values_, '\n') #собственные числа
```

```python
print_pca(pca)
```

При понижении размерности до 2, мы видим, что данные имеют около 63.5% информации. 
При уменьшении размерности до 4, данные имеют около 85% информации

```python
def show_pca_scatters(data):
    plt.scatter(data[:, 0], data[:, 1], c=labels, cmap='hsv')
    plt.show()
```

```python
show_pca_scatters(data)
```

```python
np.sum(pca.singular_values_**2) / np.sum(PCA(n_components = data.shape[1]).fit(data).singular_values_**2)
```

```python
transform_data = pca.inverse_transform(pca_data)

print('Var', np.var(data,axis=0))
print('Transform var', np.var(transform_data,axis=0),'\n')
print('Mean', np.mean(data,axis=0))
print('Transform mean', np.mean(transform_data,axis=0))
show_scatters(transform_data)
```

После восстановления данных для 4 компонент мы видим, что восстановленные данные немного отличаются от исходных. 
Отличия вызваны потерей приблизительно 15% информации. Также из-за потери информации немного отличаюется дисперсия.

```python
#svd_solver
pca_svd_solver_full = PCA(n_components = 4, svd_solver = 'full')
pca_svd_solver_full_data = pca_svd_solver_full.fit_transform(data)

pca_svd_solver_arpack = PCA(n_components = 4, svd_solver = 'arpack')
pca_svd_solver_arpack_data = pca_svd_solver_arpack.fit_transform(data)

pca_svd_solver_randomized = PCA(n_components = 4, svd_solver = 'randomized')
pca_svd_solver_randomized_data = pca_svd_solver_randomized.fit_transform(data)
```

```python
print_pca(pca_svd_solver_full)
print_pca(pca_svd_solver_arpack)
print_pca(pca_svd_solver_randomized)
```

```python
show_pca_scatters(pca_svd_solver_full_data)
```

```python
show_pca_scatters(pca_svd_solver_arpack_data)
```

```python
show_pca_scatters(pca_svd_solver_randomized_data)
```

Различные методы решения SVD дают один и тот же результат для имеющегося набора данных.

## Модификации метода главных компонент

### KernelPCA

```python
def kernel_pca_data(kernel = ''):
    kernel_pca = KernelPCA(n_components = 4, kernel = kernel)
    kernel_pca_data = kernel_pca.fit_transform(data)
    kernel_pca_all = KernelPCA(n_components=data.shape[1], kernel=kernel).fit(data)
    
    plt.title(kernel)
    show_pca_scatters(kernel_pca_data)
    
    print('Eigenvals', kernel_pca.lambdas_)
    print('Var', np.sum(kernel_pca.lambdas_) / np.sum(kernel_pca_all.lambdas_))
```

```python
kernel_pca_data(kernel='linear')
```

```python
kernel_pca_data(kernel='poly')
```

```python
kernel_pca_data(kernel='rbf')
```

```python
kernel_pca_data(kernel='sigmoid')
```

```python
kernel_pca_data(kernel='cosine')
```

При применении различных ядерных функций собственные значения ядерных матриц различны, а обощенная дисперсия имеет небольшие различия.

### SparcePCA

```python
sparse_pca = SparsePCA(n_components=2, method='lars')
sparse_pca_data = sparse_pca.fit_transform(data)

show_pca_scatters(sparse_pca_data)
```

```python
sparse_pca = SparsePCA(n_components = 2, method = 'lars', alpha = 0)
sparse_pca_data = sparse_pca.fit_transform(data)

show_pca_scatters(sparse_pca_data)
```

```python
sparse_pca = SparsePCA(n_components=2, method='lars', alpha = 0.25)
sparse_pca_data = sparse_pca.fit(data).transform(data)
plt.title('alpha = 0.25')
show_pca_scatters(sparse_pca_data)
```

SparsePCA извлекает набор разреженных компонентов. Чем выше параметр, контролирующий разреженность, тем компоненты более разреженные.
